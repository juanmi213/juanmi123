```python
import numpy as np
vector = np.arange(20) # rango del 1 al 20
vector1=np.random.rand(1,20) # que se llene de valores aleatorias
print(vector)
print(vector1)
vector2=np.random.randint(1,11,100) # llena numeros aleatorios del 1 al 11 pero de un rango del 1 al 100 como si fuera un baloto
print("aleatoria enteros",vector2)
print("tamaño",vector2.shape)
# suma de vectores
resultadovector = vector + vector1
print("suma de vectores", resultadovector)
#multiplicacion de vectores
resultadovector1 = vector * vector1
print("multiplicacion de vectores", resultadovector1)
#divicion de vectores
resultadovector2 = vector / vector1
print("Divicion de vectores", resultadovector2)
```
